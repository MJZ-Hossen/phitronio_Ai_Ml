# Question 01: [ Marks 10 ]

What is a perceptron?
Explain its three main components and their roles.

**Write Answer 01:** Perceptron learning is a type of supervised learning algorithm used in neural networks. It's a simple, single-layer neural network model that learns to classify inputs into two categories.

 It makes a prediction, compares it to the correct answer, and if it's wrong, it adjusts its weights to fix it. It keeps doing this until it stops making errors or until you accept that it can't, because the data isn't linearly separable.

Here are the three main components:
1. Inputs and Weights:  The inputs are the features or data points coming in.The weight determine how important each input is. The perceptron learns by adjusting these.

2. Summation Function: The perceptron multiplies each input by its weight and adds them all up plus a bias term. The bias is just a constant that shifts the result.

3. Activation Function: This decides the output (usually a threshold function, like output 1 if sum > 0, else 0). The idea is to turn a raw number into a meaningful signal.

What is a decision boundary in a perceptron?
Explain how the equation
w₁x₁ + w₂x₂ + b = 0
represents a decision boundary geometrically.

## Write Answer 02:
A decision boundary is the line or plane, or hyperplane that separates two classes. Everything on one side gets labeled 0, everything on the other side gets labeled 1. The perceptron's entire job is to find this boundary.

#w₁x₁ + w₂x₂ + b = 0

Here x₁ and x₂ are inputs in a 2D plot .
The equation above is the equation of a straight line on that 2D plane. That line is called decision boundary.

w₁ and w₂ are weights that control the orientation of the line. They determine which direction the line is tilted. We need to change the weights to rotate the line.

b — the bias controls the position of the line. It shifts the line left, right, up or down without rotating it. Without bias, the line is forced to pass through the origin

That's how the perceptron decides which class a point belongs to.

Why is the perceptron called a linear classifier?
Explain your answer using the concept of linear separability and logical gate examples such as AND, OR, and XOR.

## Write Answer 03:
A Perceptron is called a linear classifier because its decision boundary is a straight line (in 2D), a plane (in 3D), or a hyperplane (in higher dimensions). It makes decisions based on a linear combination of its inputs.

The equation w₁x₁ + w₂x₂ + b = 0 is the equation of a straight line. Then applies the step function: 1 if z > 0, else 0. That means all points on one side of this line get classified as 1, the other side as 0. That's the definition of a linear decision boundary.

**Linear separability:**

 It refers to a situation where two classes of points can be divided entirely by a linear boundary.

The Perceptron is mathematically designed to find this line. If the data is not linearly separable, the Perceptron algorithm will never converge—it will just keep oscillating its weights forever, trying to find a line that doesn't exist.

# Question 04: [ Marks 15 ]

Write the perceptron weight update rule and explain it intuitively.
Why are the weights updated only when the prediction is incorrect?

## Write Answer 04:
The Weight Update Rule:

$$w_{new} = w_{old} + \eta \cdot (y - \hat{y}) \cdot x_i$$
Here,

$w_{new}$: The updated weight.

$\eta$ (Eta): The Learning Rate. This controls how big or small a step we take. If it is too big, the perceptron overcorrects and bounces around and; if it is too small, the preceptron learns very slowly.

$y$: The Target Value.

$\hat{y}$: The Predicted Value.

$x_i$: The Input Value associated with that weight.

*Suppose I need to solve a simple math: x+y=20*

Now, push a value to sum them and check with the output; if the output is correct no changes but if it is wrong ,it need to addjust according to the error until the output matches. If it is a big error, it need to be changed aggressively otherwise there might be a small change.

That's the perceptron. Every wrong prediction is a miss. The update rule is the correction. The learning rate is how aggressively  we adjust after each miss.

<br>

**The weights updated only when the prediction is incorrect:**

**$(y - \hat{y})$; the error has three possible values:**

>0 = prediction was correct. No update happens.

>+1 = predicted 0, actual was 1. Weights need to increase.

>-1 = predicted 1, actual was 0. Weights need to decrease.

If the data is NOT linearly separable — mistakes never stop, updates never stop, the boundary keeps shifting forever. The perceptron never converges. It loops infinitely. That's how you know your problem is unsolvable with a single perceptron.

If the data is linearly separable. Every update moves the boundary closer to correct. Eventually there are no more mistakes. So,no more updates and training stops. Although changing them would fix something that isn't broken, resulting in unstable weight and incorrect predictions again.

That's why the weights updated only when the prediction is incorrect.

# Question 06: [ Marks 20 ]

What is meant by the “perceptron Learning Rule” or weight adjustment method?
Why can adjusting one weight affect the classification of other points?

## Write Answer 06:
The perceptron learning rule is a method used to update the weights of the model when a prediction is incorrect.

The rule is:
w_new = w_old + η · (y - ŷ) · xᵢ

Here,

wnew: The updated weight.

η (Eta): The Learning Rate. This controls how big or small a step we take. If it is too big, the perceptron overcorrects and bounces around and; if it is too small, the preceptron learns very slowly.

y: The Target Value.

y^: The Predicted Value.

xi: The Input Value associated with that weight.

When it's right, weights freeze, but updating correct predictions would destabilize what's already working. This is because the perceptron doesn’t learn separately for each point. It learns one global decision boundary.

If the data is non-linearly separable, the perceptron cannot find a single linear decision boundary that correctly classifies all points. As a result, it keeps updating the weights whenever it encounters a misclassified point.

So, changing a weight alters this boundary, which may improve the classification of some points but can also change the classification of others.
