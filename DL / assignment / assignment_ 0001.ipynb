# Question 01: [ Marks 10 ]

What is a perceptron?
Explain its three main components and their roles.

**Write Answer 01:** Perceptron learning is a type of supervised learning algorithm used in neural networks. It's a simple, single-layer neural network model that learns to classify inputs into two categories.

 It makes a prediction, compares it to the correct answer, and if it's wrong, it adjusts its weights to fix it. It keeps doing this until it stops making errors or until you accept that it can't, because the data isn't linearly separable.

Here are the three main components:
1. Inputs and Weights:  The inputs are the features or data points coming in.The weight determine how important each input is. The perceptron learns by adjusting these.

2. Summation Function: The perceptron multiplies each input by its weight and adds them all up plus a bias term. The bias is just a constant that shifts the result.

3. Activation Function: This decides the output (usually a threshold function, like output 1 if sum > 0, else 0). The idea is to turn a raw number into a meaningful signal.
