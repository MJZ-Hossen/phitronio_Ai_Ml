# Question 01: [ Marks 10 ]

What is a perceptron?
Explain its three main components and their roles.

**Write Answer 01:** Perceptron learning is a type of supervised learning algorithm used in neural networks. It's a simple, single-layer neural network model that learns to classify inputs into two categories.

 It makes a prediction, compares it to the correct answer, and if it's wrong, it adjusts its weights to fix it. It keeps doing this until it stops making errors or until you accept that it can't, because the data isn't linearly separable.

Here are the three main components:
1. Inputs and Weights:  The inputs are the features or data points coming in.The weight determine how important each input is. The perceptron learns by adjusting these.

2. Summation Function: The perceptron multiplies each input by its weight and adds them all up plus a bias term. The bias is just a constant that shifts the result.

3. Activation Function: This decides the output (usually a threshold function, like output 1 if sum > 0, else 0). The idea is to turn a raw number into a meaningful signal.

# Question 04: [ Marks 15 ]

Write the perceptron weight update rule and explain it intuitively.
Why are the weights updated only when the prediction is incorrect?

## Write Answer 04:
The Weight Update Rule:

$$w_{new} = w_{old} + \eta \cdot (y - \hat{y}) \cdot x_i$$
Here,

$w_{new}$: The updated weight.

$\eta$ (Eta): The Learning Rate. This controls how big or small a step we take. If it is too big, the perceptron overcorrects and bounces around and; if it is too small, the preceptron learns very slowly.

$y$: The Target Value.

$\hat{y}$: The Predicted Value.

$x_i$: The Input Value associated with that weight.

*Suppose I need to solve a simple math: x+y=20*

Now, push a value to sum them and check with the output; if the output is correct no changes but if it is wrong ,it need to addjust according to the error until the output matches. If it is a big error, it need to be changed aggressively otherwise there might be a small change.

That's the perceptron. Every wrong prediction is a miss. The update rule is the correction. The learning rate is how aggressively  we adjust after each miss.

<br>

**The weights updated only when the prediction is incorrect:**

**$(y - \hat{y})$; the error has three possible values:**

>0 = prediction was correct. No update happens.

>+1 = predicted 0, actual was 1. Weights need to increase.

>-1 = predicted 1, actual was 0. Weights need to decrease.

If the data is NOT linearly separable — mistakes never stop, updates never stop, the boundary keeps shifting forever. The perceptron never converges. It loops infinitely. That's how you know your problem is unsolvable with a single perceptron.

 If the data is linearly separable. Every update moves the boundary closer to correct. Eventually there are no more mistakes. So,no more updates and training stops. Although changing them would be fixing something that isn't broken resulting in unstable weight and incorrect predictions again.

That's why the weights updated only when the prediction is incorrect.
